```python
#Importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#5 Visualize the data using R/Python by plotting the graphs for assignment no. 1 and 2. Consider a suitable data set.
#a) Use Scatter plot, bar plot, Box plot and Histogram
sales = pd.read_excel("/content/Customers.xlsx")
sales.head()
```

```python
plt.figure(figsize=(4, 4))
sns.countplot(x="Gender", data =sales)
```

```python
sns.boxplot(y="Age",data=sales,hue="Profession",width=0.5)
```

```python
plt.hist(sales['Work Experience'],align='mid',histtype='bar')
plt.xlabel('Work Experience')
plt.show()
```

```python
plt.scatter(sales['Spending Score (1-100)'],sales['Work Experience'])
plt.xlabel('Spending Score (1-100)')
plt.ylabel('Work Experience')
plt.title("Scatter plot")
plt.show()
```

```python
sns.relplot(data=sales, x="Profession", y="Work Experience",kind="line",ci=None)
```

```python
sns.pointplot(x="Family Size", y="Age", data=sales)
plt.show()
```

```python
sns.jointplot(x='Family Size', y='Spending Score (1-100)', data=sales, kind='kde')
```
